### Problem 1: Define a function 'fib' that takes a number, 'n', as a parameter and prints all the Fibonacci numbers less than 'n' to the screen.

```python
def fib(n):
    
    """    
    This is a function that takes in a number as a parameter and prints a 
    list of all Fibonacci numbers that are less than the input parameter.
    
    Parameter: 
    n = integer value that is non-negative
    
    
    The golden Fibonacci sequence is defined:
    F_n = F_(n-1) + F_(n-2)
    The initial values of the sequence are F_0 = 0 and F_1 = 1.
    Hence the beginning sequence can be defined 0, 1, 1, 2, 3, 5, 8, ...
    where the sum of the previous two numbers results in the following number.
    The individual terms will result: fib(0) = 0, fib(1) = 1, fib (2) = 1, fib(3) = 2, ...
    So for taking the Fibonacci numbers "less than n" for fib(3) = [0, 1, 1] as these 
    are all numbers that preceed n = 3.
    
    Here I am taking the statement "all the Fibonacci numbers less than n" to mean all 
    numbers less than (strictly less than) the nth term indexed in the Fibonacci sequence. 
    
    ak-2015
    
    """
    
    # First, to check that parameter n is non-negative:
    if n > 0:
    
        fib_list = [0]  # Initializes an empty list to append Fibonacci numbers to.

        fib1, fib2 = 0, 1 # The intial values of the Fibonacci sequence: 0 and 1.
        
        # For loop to iterate through the range n-1 since we want to print all values LESS THAN n.
        for i in range(n-1):
            fib1, fib2 = fib2, fib1 + fib2 # Updates fib1 and fib2 per the Fibonacci sequence recurrsion.
            fib_list.append(fib1) # Appends values from for loop to fib_list. 
        print(fib_list)
    
    else:
        print('Please re-enter a parameter that is greater than or equal to 0.')

## Test:
# print(Test 1)
# fib(3)
# print('\n')


# print(Test 2)
# fib(-1)
    
```

### Problem 7: Define a function sum() and a function multiply that sums and multiplies (respectively) all the numbers in a list of numbers. For example, sum([1, 2, 3, 4]) returns 10, and multiply ([1, 2, 3, 4]) returns 24.

```python
def sum(obj):
    
    """
    The sum() function takes in a list of numbers and sums the numbers in the list.
    
    Parameter: 
    obj = list
    
    ak-2015
    
    """
    
    # First check parameter is a list:
    if type(obj) == list:
        
        list_sum = 0 # Initializes the sum of the list.
        
        # For loop to iterate through the values in the list.
        for i in obj:
            list_sum = list_sum + i # Adds the i-th value of the list to the list_sum.
        print('Sum =', list_sum)        
        
    else:
        print("Please re-enter parameter that is of type 'list'.")
        

def multiply(obj):
      
    """
    The multiply() function takes in a list of numbers and multiplies the numbers
    in the list.
    
    Parameter: 
    obj = list
    
    ak-2015
    
    """  
    
    # First check parameter is a list:
    if type(obj) == list:
        
        list_mult = 1 # Initiates the multiplication of the list, when multiplying initial value = 1 (not 0).
        
        # For loop to iterate through the list.
        for i in obj:
            list_mult = list_mult * i # Multiplies the i-th value of the list to the list_mult.
        print('Multiply =', list_mult)        
        
    else:
        print("Please re-enter parameter that is of type 'list'.")    
        
## Test:
# print('Test 1')
# sum([1,2,3,4]) # This appears that it is call the python built-in "summ()" function
# multiply([1,2,3,4])
# print('\n')

# print('Test 2')
# sum('123')
# multiply('123')
```

### Problem 8: Define a function reverse() that computes the reversal of a string. For example, reverse("I am testing") should return the string "gnitset ma I". 

```python
def reverse(string):
    
    """
    This function computes the reversal of a string. 
    For example, reverse("I am testing") should return the string "gnitset ma I". 
    
    Parameter: 
    string = string
    
    ak-2015
    
    """
    
    # Check parameter is a string:
    if type(string) == str:
        reverse_string = string[::-1] # The :: will take all values of the string from beginning to end,
                                      # and -1 will take the indexing in reverse from end to beginning.     
        print(reverse_string)
    
    else: 
        print("Please re-enter parameter that is of type 'string'.")
        

## Test:
# print('Test 1')
# reverse('I am testing')
# print('\n')

# print('Test 2')
# reverse([1, 1, 2, 3])
    
```

### Problem 9:  Define a function is_palindrome() that recognizes palindromes. For example, is_palindrome("radar") should return True.

```python
def is_palindrome(string):
    
    """
    This function recognizes palindromes. For example, is_palindrome("radar") should return True.
    
    Parameter: 
    string = string
    
    ak-2015
    
    """
    
    # Check parameter is a string:
    if type(string) == str:
        reverse_string = string[::-1] # Using code from problem 8 to find the reverse, since palindromes are the spelled
                                      # the same forwards and backwards.
        
        if string == reverse_string: # Checks if the original (forwards) string equals the reverse (backwards) string.
            print (True)
        
        else:
            print (False)
        
    else:
        print("Please re-enter parameter that is of type 'string'.")
        
## Test
# print('Test 1')
# is_palindrome('radar')
# print('\n')

# print('Test 2')
# is_palindrome('math')
# print('\n')

# print('Test 3')
# is_palindrome([1,2,2])
```

### Problem 10: Write a function is_member() that takes a value (i.e. a number, string, etc) x and a list of values a, and returns True if x is a member of a, False otherwise. 

```python
def is_member(x, a):
    
    """
    This function takes a value (i.e. a number, string, etc) x and a list of values a,
    and returns True if x is a member of a, False otherwise. 
    
    Parameters: 
    x = any value (number, string, etc.) 
    a = list
    
    ak-2015
    
    """
    
    # Check parameter 'a' is a list:
    if type(a) == list:
        
        # Can use the 'in' operator, but for this exercise will check by using a for loop.
        count = 0 # Initialize a counter
        
        for i in range(0, len(a)): # Use len(a) to determine for how many values to iterate through.
            
            if x == a[i]:                
                count += 1 # If for the case that x equals a value in list 'a', the count will increment.
            
            else:
                count = count # If False, count will remain the same. When iterating and checking each value in 
                              # the entire list, if x is never equal to a value in the list 'a', then the count will 
                              # remain 0. 
                
        # Checking the value of count. If value > 0, will print True. If value = 0, will print False.       
        if count > 0:
            print('True')
        else:
            print('False')
    
    else:
        print("Please re-enter parameter that is of type 'list'.")  

       
        
## Test
# print('Test 1')
# is_member(4, [0, 1, 2, 3])
# print('\n')

# print('Test 2')
# is_member(2, [0, 1, 2, 3])
# print('\n')

# print('Test 3')
# is_member('a', ['a', 'b'])
        
```

### Problem 11: Define a function overlapping() that takes two lists and returns True if they have at least one member in common, False otherwise. Can use is_member() function, or the in operator, you should write it using two nested for loops. 

```python
def overlapping(list1, list2):
    
    """
    This function takes two lists and returns True if they have at least one member in common,
    False otherwise. 
    
    Parameters: 
    list1 = list 
    list2 = list
    
    ak-2015
    
    """    
    
    # Check parameters are both lists:
    if (type(list1) == list) and (type(list2) == list):
        
        count = 0 # Initializes counter.
        
        # Create a nested for loop:
        for i in range(0, len(list1)): # Iterates over the range of all values in list1.
            for j in range(0, len(list2)): # Iterates over the range of all values in list2.
                if list1[i] == list2[j]: # Checks if values in list1 and list2 are equal.
                    count += 1 # If there exists an equivalence, increments the counter.
                
                else:
                    count = count # If value in list1 does not equal value in list2, counter remains the same.

        # Checking the value of count. If value > 0, will print True. If value = 0, will print False.         
        if count > 0:
            print('True')
        else:
            print('False')                   
        
    
    else:
        print("Please re-enter parameters that are of type 'list'.")   
        
## Test
# print('Test 1')
# overlapping([1,2], [1,3])
# print('\n')

# print('Test 2')
# overlapping([1,2], [3,4])
# print('\n')

# print('Test 3')
# overlapping('ab', 'ab')
```
